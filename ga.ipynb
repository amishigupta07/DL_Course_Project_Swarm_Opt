```python
import numpy as np

def sphere(x):
    return np.sum(x**2)

def rastrigin(x):
    return np.sum(x**2 - 10*np.cos(2*np.pi*x) + 10)

def rosenbrock(x):
    return np.sum(100*(x[1:] - x[:-1]**2)**2 + (x[:-1] - 1)**2)

def ackley(x):
    return -20*np.exp(-0.2*np.sqrt(np.mean(x**2))) - np.exp(np.mean(np.cos(2*np.pi*x))) + 20 + np.e

POP_SIZE = 30
DIM = 5
GENERATIONS = 100
CROSSOVER_RATE = 0.8
MUTATION_RATE = 0.05
BOUNDS = (-5, 5)

def initialize():
    return np.random.uniform(BOUNDS[0], BOUNDS[1], (POP_SIZE, DIM))

def fitness(pop, func):
    return np.array([1/(1 + func(ind)) for ind in pop])

def selection(pop, fit):
    probs = fit / np.sum(fit)
    indices = np.random.choice(len(pop), size=len(pop), p=probs)
    return pop[indices]

def crossover(p1, p2):
    if np.random.rand() < CROSSOVER_RATE:
        point = np.random.randint(1, DIM)
        return np.concatenate([p1[:point], p2[point:]])
    return p1

def mutate(ind):
    for i in range(DIM):
        if np.random.rand() < MUTATION_RATE:
            ind[i] += np.random.normal(0, 0.1)
    return ind

def GA(func):
    pop = initialize()
    best_vals = []

    for _ in range(GENERATIONS):
        fit = fitness(pop, func)
        new_pop = selection(pop, fit)

        next_gen = []
        for i in range(0, POP_SIZE, 2):
            p1, p2 = new_pop[i], new_pop[i+1]
            c1 = mutate(crossover(p1, p2))
            c2 = mutate(crossover(p2, p1))
            next_gen.extend([c1, c2])

        pop = np.array(next_gen)
        best_vals.append(np.min([func(ind) for ind in pop]))

    return best_vals[-1]

functions = [sphere, rastrigin, rosenbrock, ackley]

results = {}
for f in functions:
    vals = [GA(f) for _ in range(5)]
    results[f.__name__] = (np.min(vals), np.mean(vals), np.std(vals))

print(results)
```
