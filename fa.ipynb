```python
import numpy as np

def sphere(x):
    return np.sum(x**2)

def rastrigin(x):
    return np.sum(x**2 - 10*np.cos(2*np.pi*x) + 10)

def rosenbrock(x):
    return np.sum(100*(x[1:] - x[:-1]**2)**2 + (x[:-1] - 1)**2)

def ackley(x):
    return -20*np.exp(-0.2*np.sqrt(np.mean(x**2))) - np.exp(np.mean(np.cos(2*np.pi*x))) + 20 + np.e

POP_SIZE = 25
DIM = 5
GEN = 100
alpha = 0.2
beta0 = 1
gamma = 1
BOUNDS = (-5, 5)

def initialize():
    return np.random.uniform(BOUNDS[0], BOUNDS[1], (POP_SIZE, DIM))

def FA(func):
    pop = initialize()

    for _ in range(GEN):
        for i in range(POP_SIZE):
            for j in range(POP_SIZE):
                if func(pop[j]) < func(pop[i]):
                    r = np.linalg.norm(pop[i] - pop[j])
                    beta = beta0 * np.exp(-gamma * r**2)
                    pop[i] = pop[i] + beta*(pop[j] - pop[i]) + alpha*(np.random.rand(DIM)-0.5)

    best = min([func(ind) for ind in pop])
    return best

functions = [sphere, rastrigin, rosenbrock, ackley]

results = {}
for f in functions:
    vals = [FA(f) for _ in range(5)]
    results[f.__name__] = (np.min(vals), np.mean(vals), np.std(vals))

print(results)
```
